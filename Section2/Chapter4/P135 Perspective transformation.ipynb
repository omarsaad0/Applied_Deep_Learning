```python
import numpy as np
import cv2
import matplotlib.image as mpimg
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
image = cv2.imread(r'C:\Users\omar saad\Desktop\SDC\Section1/15.jpg')
```

```python
plt.imshow(image,cmap='gray')
```

```python
cv2.imshow('original image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
height, width = image.shape[:2]
print(width, height)
```

```python
source_points = np.float32([[117,13], [371,82], [377,226], [88,196]])
```

```python
destaniation_points = np.float32([[0,0], [width,0], [width,height], [0,height]])
```

```python
M = cv2.getPerspectiveTransform(source_points, destaniation_points)
```

```python
warped = cv2.warpPerspective(image,M,(width,height))
```

```python
cv2.imshow('warped image', warped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
croped = image[0:300, 0:300]
cv2.imshow("croped image",croped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.ones((3,3), np.uint8)
image_erosion = cv2.erode(image, kernel, iterations = 2)
image_dilation = cv2.dilate(image, kernel, iterations = 1)
```
